```python
from langchain_community.document_loaders import PyMuPDFLoader
loader=PyMuPDFLoader(file_path="genai.pdf")
```

```python
pages=loader.load()
```

```python
len(pages)
```

```python
print(pages[0].page_content)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(separators=["\n\n","\n"," ",""],
                                        chunk_size=500,
                                        chunk_overlap=75)
```

```python
chunks=[splitter.create_documents(texts=[page.page_content],metadatas=[page.metadata]) for page in pages]
```

```python
len(chunks)
```

```python
len(chunks[0])
```

```python
chunks
```

```python
chunks[0][1]
```

```python
from langchain_openai import OpenAIEmbeddings
embeddor=OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
from langchain_chroma import Chroma
db=Chroma(collection_name="ai_database",
          persist_directory="./chroma_db",
          embedding_function=embeddor)
```

```python
i:int=0
for chunk in chunks:
  ids=["doc_"+str(i)+str(j) for j in range(len(chunk))]
  db.add_documents(documents=chunk,ids=ids)
  i += 1

```
